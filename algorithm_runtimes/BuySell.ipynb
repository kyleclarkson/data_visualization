```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in data
data_location = '../src_files/'
naive = pd.read_csv('../src_files/naive.csv')
dnc = pd.read_csv('../src_files/dnc.csv')
dp = pd.read_csv('../src_files/dp.csv')
```

```python
# gather plot data

x = [value for value in naive.columns]
y_naive = naive.mean()
sd_naive = naive.std()/2

y_dnc = dnc.mean()
sd_dnc = dnc.std()/2

y_dp = dp.mean()
sd_dp = dp.std()/2
```

```python
## Plot data 

# plt.plot(x, y_naive, 'rx', label='Naive')
# plt.plot(x, y_dnc, 'bs', label='DNC')
# plt.plot(x, y_dp, 'g+', label='DP')


plt.errorbar(x,
             y_naive,
             yerr=sd_naive,
             lw=1,
             capsize=3,
             fmt='rx',
             label='Naive')
plt.errorbar(x,
             y_dnc,
             yerr=sd_dnc,
             lw=1,
             capsize=3,
             fmt='bs',
             label='DNC')
plt.errorbar(x,
             y_dp,
             yerr=sd_dp,
             lw=1,
             capsize=3,
             fmt='g^',
             label='DP')

# Set scale of axis
plt.yscale('log', nonposy='clip')

# Set labels
plt.xlabel('Problem size n')
plt.ylabel('Runtime seconds')
plt.title('Runtime of various algorithms for BuySell Problem')

# Set legend
plt.legend(loc='upper left')

plt.show()
```
