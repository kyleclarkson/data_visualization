```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Read in data 
data_location = '1000000.csv'
df = pd.read_csv(data_location)
x = df['number']
y = df['rep']
```

```python
# Plot data
# Set figure labels and size
plt.figure(figsize=(20, 8))
plt.title("Minimum Number of Ones Needed to Express Integer Using Addition and Multiplication Operators")
plt.xlabel("Integer n")
plt.ylabel("Minimum number of ones needed to express n")

# Set ticks
plt.xticks(np.arange(0, 1_000_001, 100_000))
plt.yticks(np.arange(0, 51, 5))

# Plot data
plt.plot(x, y)
# plt.savefig('out.png')
```

```python
# Define log function with params a,b to fit data to function.
def log_func(x, a, b):
    return a + b*np.log(x)
```

```python
popt, pcov = curve_fit(log_func, x, y)
```

```python
# Plot data
# Set figure labels and size
plt.figure(figsize=(20, 8))
plt.title("Minimum Number of Ones Needed to Express Integer Using Addition and Multiplication Operators")
plt.xlabel("Integer n")
plt.ylabel("Minimum number of ones needed to express n")

# Set ticks
plt.xticks(np.arange(0, 1_000_001, 100_000))
plt.yticks(np.arange(0, 51, 5))

# Plot data
plt.plot(x, y)
plt.plot(x, log_func(x, *popt), 'r-')
```
