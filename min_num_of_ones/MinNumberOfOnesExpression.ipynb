```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in data 
data_location = '1000000.csv'
df = pd.read_csv(data_location)
```

```python
# Plot data
# Set figure labels and size
plt.figure(figsize=(20, 8))
plt.title("Minimum Number of Ones Needed to Express Integer Using Addition and Multiplication Operators")
plt.xlabel("Integer n")
plt.ylabel("Minimum number of ones needed to express n")

# Set ticks
plt.xticks(np.arange(0, 1_000_001, 100_000))
plt.yticks(np.arange(0, 51, 5))

# Plot data
plt.plot(df['number'], df['rep'])
# plt.savefig('out.png')
```
